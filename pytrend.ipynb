```python
from pytrends.request import TrendReq
import pandas as pd
import gspread
from oauth2client.service_account import ServiceAccountCredentials

json_keyfile = "/Users/zzzhao/Desktop/Google trend/superb-ethos-451521-r8-184d4d1be3bb.json"

scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
creds = ServiceAccountCredentials.from_json_keyfile_name(json_keyfile, scope)
client = gspread.authorize(creds)

spreadsheet = client.open_by_url("https://docs.google.com/spreadsheets/d/1-_ld_QqNi21j7UpUKt2hg0ATpGboBFJalGNGkmZ5wCE/edit?usp=sharing")
sheet = spreadsheet.sheet1
pytrends = TrendReq(hl='en-US', tz=360)

categories = {
    "Sports": ["NHL", "NFL", "NBA", "FIFA"],
    "Anime1": ["Naruto", "One Piece", "Dragon Ball", "Jujutsu Kaisen"],
    "Anime2": ["Attack on Titan", "Spy X Family", "My Hero Academia", "Demon Slayer"],
    "Disney": ["Mickey Mouse", "Marvel", "Star Wars", "Pixar"],
    "Music": ["Kendrick Lamar", "Drake", "Taylor Swift", "Bruno Mars"],
    "Video Games 1": ["Nintendo", "Electronic Arts", "Ubisoft", "Blizzard"],
    "Video Games 2": ["Square Enix", "Riot", "Tencent", "Epic Games"],
    "Movies & TV": ["Netflix", "Disney+", "HBO", "MGM"],
    "Pop": ["Funko", "Good Smile Company", "Bandai", "Mattel"],
}

sheet.clear()

current_row = 1

for category, keywords in categories.items():
    keywords = keywords[:4]

    pytrends.build_payload(kw_list=keywords, geo='US', timeframe='now 1-d')
    df = pytrends.interest_over_time()

    if 'isPartial' in df.columns:
        df = df.drop(columns=['isPartial'])

    if not df.empty:
        latest_trend = df.iloc[-1]

        sheet.update(
            f"A{current_row}",
            [[f"{category} Section"]],
            value_input_option="USER_ENTERED"
        )
        current_row += 1


        header = ["Date"] + keywords
        sheet.update(
            f"A{current_row}",
            [header],
            value_input_option="USER_ENTERED"
        )
        current_row += 1


        date_str = df.index[-1].strftime("%Y-%m-%d")
        row_data = [date_str] + latest_trend.tolist()
        sheet.update(
            f"A{current_row}",
            [row_data],
            value_input_option="USER_ENTERED"
        )
        current_row += 2

print("âœ… Google Sheets updated successfully!")
```
